# Test highconf processing script

```python
import pandas as pd
from pathlib import Path

# ==========================================
# CONFIGURATION
# ==========================================
column_map = {
    'cdr3s_aa': 'junction_aa',
    'clonotype_id': 'clonotype_id',
    'count': 'count',
    'v_call': 'v_call',
    'd_call': 'd_call',
    'j_call': 'j_call'
}

# Set up path and filename
path = Path("/Users/ishaharris/Projects/TCR/TCR-Isha/data/Count_Matrices")
f = "AC02_countmatrix_ng.tsv"
n = 10

# Read the file
cm = pd.read_csv(path / f, sep='\t')

# ==========================================
# PART 1: FILTERING
# ==========================================
col_of_interest = ['positive']
cols_to_remove = [col for col in cm.columns if col.startswith('peptide')]

# Mask 1: keep rows where col of interest > n
mask_keep_interest = (cm[col_of_interest] > n).all(axis=1)

# Mask 2: keep rows where ALL peptide col < n
mask_remove_high_peptides = (cm[cols_to_remove] <= n).all(axis=1)

# Apply filters
cm_filtered = cm[mask_keep_interest & mask_remove_high_peptides].copy()

# ==========================================
# PART 2: EXPLODE TRB LOGIC (Optimized)
# ==========================================
print("Exploding and filtering TRB sequences...")

# 1. Split the string by ';' into a list
#    Example: "TRA:XYZ;TRB:ABC" -> ["TRA:XYZ", "TRB:ABC"]
cm_filtered['cdr3s_aa'] = cm_filtered['cdr3s_aa'].str.split(';')

# 2. Explode the list
#    This creates a cloned row for every item in the list automatically
cm_exploded = cm_filtered.explode('cdr3s_aa')

# 3. Filter: Keep only rows where the specific exploded CDR3 starts with "TRB:"
#    This effectively deletes the TRA rows created by the explode
cm_exploded = cm_exploded[cm_exploded['cdr3s_aa'].str.startswith('TRB:', na=False)]

# 4. Clean: Remove the "TRB:" prefix from the string
cm_exploded['cdr3s_aa'] = cm_exploded['cdr3s_aa'].str.replace('TRB:', '', regex=False)

print(f"Shape after TRB explosion: {cm_exploded.shape}")

cm_exploded['count'] = cm_exploded['positive']

# ==========================================
# PART 3: APPLY MAPPING (SELECT & RENAME)
# ==========================================

# 1. Filter the map: Ensure keys exist
valid_map = {k: v for k, v in column_map.items() if k in cm_exploded.columns}

# 2. Select keys and Rename
cm_final = cm_exploded[list(valid_map.keys())].copy()
cm_final.rename(columns=valid_map, inplace=True)

# Optional: View the result
print(cm_final.head())
```
